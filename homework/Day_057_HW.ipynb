# 作業
- 作業目標 : 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
y_pred = AgglomerativeClustering(n_clusters=3, linkage='ward').fit_predict(X[:, :2])
a = y_pred == 0
b = y_pred == 1
c = y_pred == 2
y_pred[a], y_pred[b], y_pred[c] = 2, 0, 1
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
y_pred = AgglomerativeClustering(n_clusters=3, linkage='complete').fit_predict(X[:, :2])
a = y_pred == 0
b = y_pred == 1
c = y_pred == 2
#y_pred[a], y_pred[b], y_pred[c] = 2, 0, 1
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
y_pred = AgglomerativeClustering(n_clusters=3, linkage='average').fit_predict(X[:, :2])
a = y_pred == 0
b = y_pred == 1
c = y_pred == 2
y_pred[a], y_pred[b], y_pred[c] = 0, 2, 1
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
y_pred = AgglomerativeClustering(n_clusters=3, linkage='single').fit_predict(X[:, :2])
a = y_pred == 0
b = y_pred == 1
c = y_pred == 2
#y_pred[a], y_pred[b], y_pred[c] = 0, 1, 2
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```
