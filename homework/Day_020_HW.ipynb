```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = '../data/Part02/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""
best_i = 0
best_j = 0
best_score = 0
for i in range(100, 5001, 100):
    for j in range(i + 1000, 5001, 100):
        tmp = df.copy()
        tmp['1stFlrSF'] = tmp['1stFlrSF'].clip(i, j)

        # 做線性迴歸, 觀察分數
        train_X = MMEncoder.fit_transform(tmp)
        estimator = LinearRegression()
        score = cross_val_score(estimator, train_X, train_Y, cv=5).mean()
        if score > best_score:
            best_score = score
            best_i = i
            best_j = j

tmp = df.copy()
tmp['1stFlrSF'] = tmp['1stFlrSF'].clip(best_i, best_j)
sns.regplot(x = tmp['1stFlrSF'][:train_num], y=train_Y)
plt.show()
print(best_i, best_j, best_score)
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""

best_i = 0
best_j = 0
best_score = 0
for i in range(100, 5001, 100):
    for j in range(i + 1000, 5001, 100):
        keep_indexs = (df['GrLivArea']> i) & (df['GrLivArea']< j)
        tmp = df[keep_indexs]
        tmp_Y = train_Y[keep_indexs]

        # 做線性迴歸, 觀察分數
        try:
            train_X = MMEncoder.fit_transform(tmp)
            estimator = LinearRegression()
            score = cross_val_score(estimator, train_X, tmp_Y, cv=5).mean()
            if score > best_score:
                best_score = score
                best_i = i
                best_j = j
        except:
            pass

keep_indexs = (df['GrLivArea']> best_i) & (df['GrLivArea']< best_j)
tmp = df[keep_indexs]
tmp_Y = train_Y[keep_indexs]
sns.regplot(x = tmp['1stFlrSF'][:train_num], y=tmp_Y)
plt.show()
print(best_i, best_j, best_score)
```
